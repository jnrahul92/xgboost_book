```python
mean_temp = (df_bikes.iloc[700]['temp'] + df_bikes.iloc[702]['temp'])/2
mean_atemp = (df_bikes.iloc[700]['atemp'] + df_bikes.iloc[702]['atemp'])/2
```

```python
df_bikes['temp'].fillna(mean_temp, inplace=True)
df_bikes['atemp'].fillna(mean_atemp, inplace=True)
```

```python
df_bikes['dteday'] = pd.to_datetime(df_bikes['dteday'], infer_datetime_format=True)
```

```python
df_bikes.columns
```

```python
X = df_bikes.iloc[:,:-1]
y = df_bikes.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
import warnings
```

```python
warnings.simplefilter("ignore")
```

## Linear Regression

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train, y_train)
```

```python
y_pred = lin_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
import numpy as np
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = mse**0.5
```

```python
print("RMSE : %0.2f" %(rmse))
```

## XGBoost

```python
from xgboost import XGBRegressor
```

```python
xg_reg = XGBRegressor()
```

```python
xg_reg.fit(X_train, y_train)
```

```python
y_pred = xg_reg.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = mse**0.5
```

```python
print("RMSE : %0.2f" %(rmse))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
model = LinearRegression()
```

```python
scores = cross_val_score(model, X, y, scoring="neg_mean_squared_error", cv=10)
```

```python
rmse = np.sqrt(-scores)
```

```python
print("Reg rmse : ",np.round(rmse,2))
print("RMSE Mean : ", round(rmse.mean(),2))
```

```python
model = XGBRegressor()
```

```python
scores = cross_val_score(model, X, y, scoring="neg_mean_squared_error", cv=10)
rmse = np.sqrt(-scores)
print("Reg rmse : ",np.round(rmse,2))
print("RMSE Mean : ", round(rmse.mean(),2))
```

```python
df_census = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",header=None)
```

```python
df_census.head()
```

```python
df_census.columns = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation',
                    'relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country',
                    'income']
```

```python
df_census.head()
```

```python
df_census.info()
```

```python
df_census.drop(columns=['education'],axis=1,inplace=True)
```

```python
df_census = pd.get_dummies(df_census)
```

```python
df_census.head()
```

```python
df_census.drop(columns=['income_ <=50K'],inplace=True)
```

```python
X = df_census.iloc[:,:-1]
y = df_census.iloc[:,-1]
```

```python
X.shape, y.shape
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
def cross_val(classifier, num_splits = 10):
    model = classifier
    scores = cross_val_score(model, X, y , cv = num_splits)
    print("Accuracy : ",np.round(scores,2))
    print("Mean Accuracy : ",np.round(scores.mean(),2))
```

```python
cross_val(LogisticRegression())
```
